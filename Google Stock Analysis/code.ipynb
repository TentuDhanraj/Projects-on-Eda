```python
# Closing price with moving averages

sorted_df = stock_data.sort_values(by = 'Date' , ascending = False) # Sort the data by date
sorted_df['Avg 20'] = sorted_df['Close'].rolling(window=20).mean() #Get Avg closing price in the 20 days
sorted_df['Avg 50'] = sorted_df['Close'].rolling(window=50).mean() #Avg closing price in the 50 days

# Plot
plt.figure(figsize= (12,6))
plt.plot(sorted_df['Date'],sorted_df['Close'], color = 'blue', label = 'Closing Price')
plt.plot(sorted_df['Date'] ,sorted_df['Avg 20'], color = 'green', label = '20 Day avg cls price')
plt.plot(sorted_df['Date'] ,sorted_df['Avg 50'], label = '50 Day avg cls price', color = 'orange')
plt.title("Moving Average price of closing price", size = 10)
plt.xlabel("Date")
plt.ylabel('Clsoing Price (USD)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Create year month coloumn
sorted_df.columns
sorted_df['Date'] = pd.to_datetime(sorted_df['Date'])
sorted_df['YearMonth'] = sorted_df['Date'].dt.to_period('M')

# Group by 'YearMonth' and calculate average Close
monthly = sorted_df.groupby('YearMonth')['Close'].mean().reset_index()
monthly['YearMonth'] = monthly['YearMonth'].astype(str) # Convert for plotting

# Calculate Bollinger Bands
monthly['MA'] = monthly['Close'].rolling(window=3).mean()
monthly['STD'] = monthly['Close'].rolling(window=3).std()
monthly['Upper'] = monthly['MA'] + 2 * monthly['STD']
monthly['Lower'] = monthly['MA'] - 2 * monthly['STD']

# Plot
plt.figure(figsize=(12, 6))
plt.plot(monthly['YearMonth'].head(50), monthly['Close'].head(50), label='Monthly Avg Close', color='blue')
plt.plot(monthly['YearMonth'].head(50), monthly['Upper'].head(50), label='Upper Band', linestyle='--', color='green')
plt.plot(monthly['YearMonth'].head(50), monthly['Lower'].head(50), label='Lower Band', linestyle='--', color='red')
plt.fill_between(monthly['YearMonth'].head(50), monthly['Upper'].head(50), monthly['Lower'].head(50), color='gray', alpha=0.2)

plt.title('Monthly Avg Close Price with Bollinger Bands')
plt.xlabel('Month')
plt.ylabel('Average Close Price (USD)')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()

#When standard deviation increases, the bands widen → indicates higher volatility.
# When standard deviation increases, the bands widen → indicates higher volatility.
# If price touches upper band: It may be overbought.
# If price touches lower band: It may be oversold.


```

```python


# Calculate daily returns
sorted_df['Daily Return'] = sorted_df['Close'].pct_change() * 100  # % returns

# Plot histogram
plt.figure(figsize=(10, 5))
sns.histplot(sorted_df['Daily Return'].dropna(), bins=50, kde=True, color='purple')
plt.title('Distribution of Daily Returns')
plt.xlabel('Daily Return (%)')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()

# If it's centered around 0 → most daily changes are small.

# Fat tails (long ends) → potential big gains or losses.

```
